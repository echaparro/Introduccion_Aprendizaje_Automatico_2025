Esta tarea consiste en calcular eigenvalores y eigenvectores de matrices siguiendo el calculo a mano, usando la librería numpy y sympy.

Los eigenvalores y eigenvectores son conceptos fundamentales en álgebra lineal y tienen aplicaciones en machine learning, física, ingeniería y más.

```python
import numpy as np
import sympy as sym
```

Sea A una matriz de nxn con componentes reales, el número λ se denomina valor característico de A, si existe un vector diferente de cero V en los numeros complejos tal que  Av = λv.

El vector V se denomina vector característico de A
Procedimiento:

1.- Se encuentra P(λ) = det(A -λI)

2.- Se encuentran las raíces de P(λ)=0

3.-Se resuelve el sistema homogeneo (A -λI)V=0



```python
def EigenVal_Vec(parMatriz):
    print(f"{"=="*10} Cálculo paso a paso {"=="*10}")
    print(f"1.- Se encuentra P(λ) = det(A -λI)")
    lambda_sym = sym.symbols('λ')
    print(f"   det(A -λI) =0")
    columns=parMatriz.shape[0]
    print(f"{"--"*10} λI {"--"*10}")
    I= np.eye(columns)
    Ilambda= lambda_sym*I
    print(f"λI = {Ilambda}")

    print(f"{"--"*10} A-λI {"--"*10}")
    AlamdaI=A-Ilambda
    print(f"A-λI = {AlamdaI}")

    print(f"{"--"*10} det(A -λI) {"--"*10}")
    res = sym.Matrix(AlamdaI)
    detRes= res.det()
    print(f" det(A -λI) = {detRes} = 0")
    soluciones = sym.solve(detRes, lambda_sym)
    print(f"2.- Se encuentran las raíces de P(λ)=0")
    print(f"  Eigen Valores: {[float(sol) for sol in soluciones]}")

    print("3.-Se resuelve el sistema homogeneo (A -λI)V=0")
    vectorCaracteristico=[]
    for i in range(columns):
        vectorCaracteristico.append(sym.symbols('X'+ str(i + 1)))

   
    for val in soluciones:
        print(f"{"--"*10} (A -λI)=V  ----- λ={float(val)} {"--"*10}")
        print(f"{parMatriz} - {float(val)} * {I} = {vectorCaracteristico}")
        res=parMatriz-(float(val) * I)
        print(f"{"--"*10} (A -λI)V=0 {"--"*10}")
        print(f"{res} = ll0")
        print(f"{res} * {vectorCaracteristico} = 0")
        print(f"{"--"*15}")
        print(f"{res * vectorCaracteristico} = 0")
        ecuaciones = res * vectorCaracteristico
        print(f"{"--"*10} Ecuacinoes {"--"*10}")
        for i in range(columns):
            ecuacionstr=""
            ecuacion=""
            for j in range(columns):
               ecuacionstr += f"{ecuaciones[i,j]} + "
            
            ecuacion = ecuacionstr.rstrip(' + ') + " = 0"
            print(f"{ecuacion}")

        U, S, Vt = np.linalg.svd(res)
        null_space = Vt[-1:]  # El último vector singular derecho

        print(f"Eigen Vector: {null_space[0]}")
```

```python

def CalculaEigenVal_Vec(parMatriz):

    print(f"{"=="*15} Numpy {"=="*15}")
    eiVal, eiVec = np.linalg.eig(parMatriz)

    print(f"Eigenvalores: {eiVal}")
    print(f"Eigenvectores: {eiVec}")


    print(f"{"=="*15} Sympy {"=="*15}")
    A = sym.Matrix(parMatriz)
    eiVal = A.eigenvals()
    eiVec = A.eigenvects()
    print(f"Eigenvalores: {eiVal}")
    print(f"Eigenvectores: {eiVec}")

    EigenVal_Vec(parMatriz)
```

```python
A = np.array([[4, 2],
              [1, 3]])

CalculaEigenVal_Vec(A)
```
